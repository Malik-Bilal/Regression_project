```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/annual_temp.csv")
data
```

```python
data_GCAG = data.loc[data.Source == "GCAG"]
data_GCAG
```

```python
data_GCAG.info()
```

```python
X = data_GCAG.iloc[:, 1:2].values
Y = data_GCAG.iloc[:, -1:].values
print("X values: \n", X)
print("Y values: \n", Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3, random_state= 0)
```

```python
print(X_train)
print(Y_train)
```

```python
# X_train = X_train.reshape(-1, 1)
# Y_train = Y_train.reshape(-1, 1)
# Y_train
print(X_train.shape)
print(Y_train.shape)
```

```python
# print(X_train)
# print(Y_train)
```

LinearRegression :

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
plt.scatter(X_train, Y_train, color="red")
plt.plot(X_train, regressor.predict(X_train), color = "blue")
plt.title("Annual temperature vs Year")
plt.xlabel("Year")
plt.ylabel("Annual temperature (Mean Value)")
```

```python
plt.scatter(X_test, Y_test, color="orange")
plt.plot(X_train, regressor.predict(X_train), color = "green")
plt.title("Annual temperature vs Year")
plt.xlabel("Year")
plt.ylabel("Annual temperature (Mean Value)")
plt.show()
```

Multiple Linear Regression:

```python
X_test = X_test.reshape(-1, 1)
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
plt.scatter(X_train, Y_train, color="red")
plt.plot(X_train, regressor.predict(X_train), color = "blue")
plt.title("Annual temperature vs Year")
plt.xlabel("Year")
plt.ylabel("Annual temperature (Mean Value)")
```

```python
plt.scatter(X_test, Y_test, color="orange")
plt.plot(X_train, regressor.predict(X_train), color = "green")
plt.title("Annual temperature vs Year")
plt.xlabel("Year")
plt.ylabel("Annual temperature (Mean Value)")
plt.show()
```

Support Vector Regression :

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_Y = StandardScaler()

X = sc_X.fit_transform(X)
Y = sc_Y.fit_transform(Y)
# print(X)
# print(Y)

# Fitting the Regression Model to the dataset
from sklearn.svm import SVR
regressor = SVR(kernel="rbf")
# regressor = SVR(kernel="linear")
# regressor = SVR(kernel="poly")
# regressor = SVR(kernel="sigmoid")
# regressor = SVR(kernel="precomputed")
regressor.fit(X,Y)

#predicting a new result
Y_pred = regressor.predict([[2016]])
print(Y_pred)
B = np.reshape(Y_pred, (1,1))
C = sc_Y.inverse_transform(B)
# print(C)

#Visualozation the regression results
plt.scatter(X, Y, color="red")
plt.plot(X, regressor.predict(X), color = "blue")
plt.title("Annual temperature vs Year")
plt.xlabel("Year")
plt.ylabel("Annual temperature (Mean Value)")
plt.show()

#Visualozation the regression results (for higher resolution and smoother curve)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X,Y, color = "red")
plt.plot(X, regressor.predict(X), color = "blue")
plt.title("Annual temperature vs Year")
plt.xlabel("Year")
plt.ylabel("Annual temperature (Mean Value)")
plt.show()
```
